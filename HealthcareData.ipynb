# EDA

# Univariate Analysis

# Gender

```python
data['Gender'].value_counts()
```

```python
ax = sns.countplot(data = data, x = 'Gender', palette = 'mako')
for bars in ax.containers:
    ax.bar_label(bars)
plt.title("Gender")
plt.show()
```

# Blood Type

```python
data['Blood Type'].value_counts()
```

```python
sns.set(rc = {'figure.figsize' : (10,5)})
ax = sns.countplot(data = data, x = 'Blood Type', palette = 'rocket')
for bars in ax.containers:
    ax.bar_label(bars)
plt.title("Blood Type")
plt.show()
```

# Medical Condition

```python
data['Medical Condition'].value_counts()
```

```python
sns.set(rc = {'figure.figsize' : (10,5)})
ax = sns.countplot(data = data, y = 'Medical Condition', palette = 'viridis')
for bars in ax.containers:
    ax.bar_label(bars)
plt.title('Medical Condition')
plt.show()
```

```python
data['Insurance Provider'].value_counts()
```

```python
ax = sns.countplot(x = 'Insurance Provider', data = data, palette = 'flare')
for bars in ax.containers:
    ax.bar_label(bars)
plt.title('Insurance Provider')
plt.show()
```

# Admission Type

```python
data['Admission Type'].value_counts()
```

```python
sns.set(rc = {'figure.figsize':(8,5)})
ax = sns.countplot(y ='Admission Type', data = data, palette = 'crest')
for bars in ax.containers:
    ax.bar_label(bars)
plt.title('Admission Type')
plt.show()
```

# Medication

```python
data['Medication'].value_counts()
```

```python
sns.set(rc = {'figure.figsize':(10,6)})
ax = sns.countplot(x = 'Medication', data = data, palette = 'magma')
plt.title('Medication')
plt.show()
```

# Test Results

```python
sns.set(rc = {'figure.figsize':(8,6)})
ax = sns.countplot(y = 'Test Results', data = data, palette = 'viridis')
plt.title('Medication')
plt.show()
```

# Medical Condition and Admission Type

```python
sns.set(rc = {'figure.figsize':(11,7)})
sns.countplot(x = 'Medical Condition', hue = 'Admission Type', data = data, palette = 'rocket_r')
plt.show()
```

# Billing amount range for each medical condition

```python
sns.set(rc = {'figure.figsize':(11,6)})
sns.boxplot(x = 'Medical Condition', y = 'Billing Amount', data = data, palette = 'cubehelix')
plt.show()
```

# Medications Count for Each Medical Condition

```python
data['Medication'].unique()
```

```python
sns.countplot(x = 'Medical Condition', hue = 'Medication', data = data, palette = 'viridis')
for bars in ax.containers:
    ax.bar_label(bars)
plt.title('Medications for Each Medical Condition')
plt.show()
```

# Gender vs. Medical Condition

```python
data.groupby(['Medical Condition', 'Gender']).size().unstack().plot(kind='bar', stacked=True)
plt.title('Gender vs. Medical Condition')
plt.show()
```

# Scatterplot for test results with medical conditions and bills

```python
sns.barplot(x='Test Results', y='Billing Amount', hue='Medical Condition', data=data)
plt.title('Test Results vs. Billing Amount by Medical Condition')
plt.show()
```

# Age vs Gender

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x = 'Gender', y = 'Age', data = data, palette = 'rocket')
plt.title('Age by Gender')
plt.show()
```

# Pairplot between Medical condition and [age, billing amount]

```python
sns.pairplot(data = data, hue='Medical Condition', vars=['Age', 'Billing Amount'])
plt.show()
```
